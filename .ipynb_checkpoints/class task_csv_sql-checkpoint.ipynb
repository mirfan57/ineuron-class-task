```python
outfile = open("carbon_tubes.csv", "w")
outfile.header = "Chi_indice_n, Chi_indice_m, Ini_atomic_u,Ini_atomic_v, Ini_atomic_w, Cal_atomic_u', Cal_atomic_v',Cal_atomic_w'\n"
outfile.write(outfile.header)

#outfile.seek(0)
inserted, not_inserted=0,0
with open('carbon_nanotubes.csv', 'r') as f:
    f_reader = csv.reader(f, delimiter=";") #csv.reader is a generator
    header = next(f_reader)

    for line in f_reader:
        #print(tuple(map(int,line[2].split(','))))
        n = line[0]
        m = line[1]
        u = tuple(map(int,line[2].split(',')))
        v = tuple(map(int,line[3].split(',')))
        w = tuple(map(int,line[4].split(',')))
        uu = tuple(map(int,line[5].split(',')))
        vv = tuple(map(int,line[6].split(',')))
        try:
            tt = tuple(map(int,line[7].split(',')))
            inserted += 1
        except Exception as e:
            print("value not found, error:",str(e))
            not_inserted += 1
        row = "{}, {}, {}, {}, {}, {}, {}, {}\n".format(n, m, u, v, w, uu, vv, tt)
        outfile.write(row)
print(inserted, not_inserted)
outfile.close() 
```

```python
import csv
import mysql.connector as connection

class csv_to_sql:
    
    def __init__(self, host, user, passwd, database, usepure):
        self.host = host
        self.user = user
        self.passwd = passwd
        self.database = database
        self.use_pure = usepure
    
    def db_connect(self):
        mydb = connection.connect(host=self.host, user=self.user, database=self.database, passwd=self.passwd, use_pure=self.use_pure)
        return (mydb.is_connected())
    
class create_table(csv_to_sql):
    
    def __init__(self):
        self.c = 20
    
    def my_table(self):
        mydb = self.mydb
        cur = mydb.cursor()
        try:
            query = "create table table1 (bike_no int, name varchar(50), team varchar(50))"
            cur.execute(query)
        except:
            print("Already exists.")
        que = "insert into table1 values (12, 'maverick', 'Aprilia')"
        cur.execute(que)
        mydb.commit()
        mydb.close()

con = csv_to_sql("localhost", "root", "mysql123", "mydb123", True)        
obj = create_table()
con.db_connect()
obj.my_table()


```
